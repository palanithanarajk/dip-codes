```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('imgs/airplane.png',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
img_1=img-40
plt.imshow(img_1,cmap='gray')
```

```python
img_2=img*0.5
plt.imshow(img_2,cmap='gray')
```

```python
img_3=255-img
plt.imshow(img_3,cmap='gray')
```

```python
(m,n)=img.shape
img_4=np.zeros([m,n])
```

```python
img.shape
```

```python
t=75
for i in range (512):
    for j in range (512):
        if img[i,j]<t:
            img_4[i,j]=0
        else:
            img_4[i,j]=255
    
```

```python
plt.imshow(img_4,cmap='gray')
```

```python
img_5=np.zeros([m,n])
```

```python
t1=110
t2=150
for i in range (512):
    for j in range (512):
        if img[i,j]>t1 and img[i,j]<t2:
            img_5[i,j]=255
        else:
            img_5[i,j]=0
```

```python
plt.imshow(img_5,cmap='gray')
```
